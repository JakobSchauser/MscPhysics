```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.1

D = .5

T = 10

N_runs = 1000

total_steps = int(T/dt)

sigma = 2*D*dt

gauss = np.random.normal(0, sigma, (total_steps, N_runs))


Xs = np.zeros((total_steps, N_runs))

Xs[0,:] = np.ones(N_runs)

for i in range(1, total_steps):
    Xs[i,:] = Xs[i-1,:] + gauss[i, :]


plt.figure(figsize=(12,10))
plt.plot(Xs)
plt.show()
```

```python
plt.plot(Xs.mean(axis=1))
plt.show()
plt.plot(Xs.var(axis=1))
```
