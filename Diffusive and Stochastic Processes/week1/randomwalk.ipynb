```julia
function step()
    if rand() < 0.5
        return -1
    end
    return 1
end

function walk(n_movers, n_steps)
    movers = zeros(Int8, (n_movers, n_steps))

    for s in 2:n_steps
        for m in 1:n_movers
            movers[m, s] = movers[m, s - 1] + step()
        end
    end

    histogram(transpose(movers))
end


walk(4,10)
```
