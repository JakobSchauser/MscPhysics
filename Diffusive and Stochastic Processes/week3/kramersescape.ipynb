```python
import numpy as np
import matplotlib.pyplot as plt

a = 0.2
b = 1

U = lambda x: -a*x**3/3 + b*x**2/2

U_p = lambda x: -a*x**2 + b*x

dt = 0.01

X_dot = lambda x, dt: -U_p(x)*dt + np.random.normal(0,np.sqrt(2*dt))

N = int(5000/dt)

x_s = np.zeros(N)
t = 0
for i in range(1,N):
    t += dt
    x_s[i] = x_s[i-1] + X_dot(x_s[i-1], dt)
    if x_s[i] > 8:
        x_s[i:] = None
        break



x_axis = np.linspace(-3,8,100)
plt.plot(x_s, np.linspace(-1,5,N))
plt.plot(x_axis, U(x_axis))
plt.ylim(-2,6)
plt.xlim(-3,8)
plt.hlines(0,-3,8, ls = '--', alpha = 0.5)
```
